```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([12,23,34,45,56])
y = np.array([15,21,37,42,59])
z = np.array([87,75,54,36,25])
```

# Lineplot

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, color = 'r', marker = 'o', mfc = 'black', ls = '--',
        lw = 2, ms = 7);
# Marker => >, <, o, ^, D, v
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, z, color = 'b', marker = 'o', mfc = 'black', ls = '--',
        lw = 2, ms = 7);
```

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, y, color = 'r', marker = 'o', mfc = 'black', ls = '--',
        lw = 2, ms = 7, label = '2022')
plt.plot(x, z, color = 'b', marker = 'o', mfc = 'black', ls = '--',
        lw = 2, ms = 7, label = '2023');
plt.legend();
plt.grid()
```

# Subplot

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x, y, color = 'r', marker = 'o', mfc = 'black', ls = '--',
        lw = 2, ms = 7, label = '2022')

plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x, z, color = 'b', marker = 'o', mfc = 'black', ls = '--',
        lw = 2, ms = 7, label = '2023');
plt.grid()
```

# Bar plot

```python
courses = ['Java','C','Python','R','Kotlin']
students = [67,70,74,45,50]
```

```python
plt.title('Student Enrollment')
plt.xlabel('Courses')
plt.ylabel('No. of Students')
mycolors = ['r', 'black','cyan', 'yellow', 'green']
plt.grid()
plt.bar(courses, students, color = mycolors, width = 0.5,
       edgecolor = 'r', linewidth = 4, ls = '--', alpha = 0.8);
```

# Scatter plot

```python
plt.title('Line Plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.scatter(x, y, marker = 'o');
```

# Pie Chart

```python
students
```

```python
courses
```

```python
plt.title('Pie Chart')
plt.pie(students, labels = courses, autopct= '%.2f%%', radius= 0.8,
       labeldistance= 1.2, shadow= True, startangle= 90);
```

# Boxplot

```python
marks = [67,63,62,69,59,66,61,70,81]
```

```python
plt.boxplot(marks);
```

```python
Q3 = np.percentile(marks, 75)
Q3
```

```python
Q1 = np.percentile(marks, 25)
Q1
```

```python
IQR = Q3 - Q1
IQR
```

```python
Upper_bound = Q3 + 1.5*IQR
Upper_bound
```

```python
Lower_bound = Q1 - 1.5*IQR
Lower_bound 
```
